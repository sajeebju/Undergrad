```python
import numpy as np
import matplotlib.pyplot as plt

array = np.linspace(-10, 10, 1000)
array_n = np.sin(array)

#Linearplot
class LinePlot:
    def __init__(self, X, Y, color, title, lw, background_style):
        self.X = X
        self.Y = Y
        self.title = title
        self.color = color
        self.lw = lw
        self.background_style = background_style
        
    
    def plot_line(self):
        plt.style.use(self.background_style)
        plt.plot(self.X, self.Y, color = self.color , lw = self.lw)
        plt.xlim(-15, 15)
        plt.box(on = None)
        plt.xticks([])
        plt.yticks([])
        plt.title(self.title)
        plt.show()
    
        
class ScatterPlot:
    def __init__(self, A, B, title, color, alpha,  cmap):
        self.A = A
        self.B = B
        self.title = title
        self.color = color
        self.alpha = alpha
        self.cmap = cmap
        
    def plot_scatter(self):
        plt.scatter(self.A, self.B, color = self.color, alpha = self.alpha,
                    cmap = self.cmap)
        
        plt.xlim(-15, 15)
        plt.ylim([])
        plt.box(on = None)
        plt.xticks([])
        plt.yticks([])
        plt.title(self.title)
        plt.show()

        
#Piechart

values = [450, 300, 290, 150, 600, 350, 100]
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
colors = ['r', 'k', 'c', 'y', 'b', 'g', 'm']
explode = [0, 0, 0, 0.2, 0, 0, 0]

class PieChart:
    def __init__(self, values, labels, colors, explode, autopct, shadow,
                 startangle, background_style, title):
        
        self.values = values
        self.labels = labels
        self.colors = colors
        self.explode = explode
        self.autopct = autopct
        self.shadow = shadow
        self.startangle = startangle
        self.background_style = background_style
        self.title = title
        
    def plot_pie(self):
        plt.style.use(self.background_style)
        plt.pie(self.values, labels = self.labels, colors = self.colors,
                explode= self.explode, autopct= self.autopct,
        shadow= self.shadow, startangle= self.startangle)
        
        plt.title(self.title)
        plt.show()
        
        
#3D_Plot  
        
x = np.linspace(-6, 6, 1000)
y = np.linspace(-6, 6, 1000)

def func_3d(p, q):
  return np.cos(p**3 + q**4)
  
X, Y = np.meshgrid(x, y)
Z = func_3d(X, Y)

class Plot_3D:
    def __init__(self, X, Y,Z, cmap):
        self.X = X
        self.Y = Y
        self.Z = Z
        self.cmap = cmap
    
    def plot_3d(self):
        
        plt.figure(figsize = (10, 10))
        ax = plt.axes(projection = '3d')
        ax.contour(self.X, self.Y, self.Z, 100, cmap = self.cmap,  extend3d=False ,
                   stride=5, zdir='z', offset=None, )
        plt.show()
        

        
#object1 = LinePlot(array, array_n, 'r', 'Normal Lineplot',  2, 'dark_background')
#print(object1.plot_line())
        
#obj_scatter = ScatterPlot(array, array_n, 'Normal Scatterplot', 'c', 0.7, 'viridis')
#print(obj_scatter.plot_scatter())

#obj_pie = PieChart(values, labels, colors, explode, "%1.1f%%" , False, 90 , 'seaborn', "A simple pie-chart" )
#print(obj_pie.plot_pie())
    
obj_3d = Plot_3D(X, Y,Z, 'viridis')
print(obj_3d.plot_3d())
```
